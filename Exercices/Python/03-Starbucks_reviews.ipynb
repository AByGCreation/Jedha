1. Import pandas

```python
import pandas as pd
from datetime import datetime
```

2. Read the data from starbucks_reviews.csv

```python
df = pd.read_csv("../resources/starbucks_reviews.csv")
display(df)
```

3. How many columns/lines are in the dataset?

```python
print(df.shape)
```

4. Convert the column "date" into datetime type.

üí° You'll have to specify the format with python's [strptime syntax](https://docs.python.org/3/library/datetime.html#strftime-strptime-behavior)

```python
df.date = df.apply(lambda x: datetime.strptime(x.date, "%B %d, %Y").timestamp(), axis=1)
display(df)
```

5. Extract some features from the "date" column:
* year
* month
* day of week

```python

def datetimeConverter(date):
    year = datetime.fromtimestamp(date).year
    month = datetime.fromtimestamp(date).month
    weekday = datetime.fromtimestamp(date).weekday()
    return (year, month, weekday)

df['year'] = df.apply(lambda x: datetimeConverter(x.date)[0], axis=1)
df['month'] = df.apply(lambda x: datetimeConverter(x.date)[1], axis=1)
df['weekday'] = df.apply(lambda x: datetimeConverter(x.date)[2], axis=1)

display(df)
```

6. Convert the "Review" column into lowercase

```python
df.Review = df.Review.str.lower()
display(df)
```

7. Extract features from the "Review" column:
* Count the occurrences of the word "starbucks"
* Count the total number of words in each review

```python
keyword = 'starbucks'
df['starRev'] = df.Review.apply(lambda x: str.count(x, keyword) if str.find(x, keyword) else None)
df['countWord'] = df.Review.apply(lambda x: len(str(x).split() ))

display(df)
```

8. Use pandas' `.describe()` method to get some basic statistics about the dataset and answer the questions below:
* Are the reviews rather good or bad?
* Are the reviews recent or old?
* Are the reviews rather long or short?
* Do people mention a lot the word "starbuck" ?

```python
df['revrat'] = round(100/(df.countWord / df.starRev),2)

display(df)

display(df['Rating'].describe())
display(df['revrat'].describe())
display(df['countWord'].describe())


display (df.describe(include='all'))




```
